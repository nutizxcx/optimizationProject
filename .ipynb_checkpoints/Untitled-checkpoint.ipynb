```python
from random import random, randrange, shuffle
import pandas

proteinIndex = 11
caloriesIndex = 4
sugarIndex = 12
carbIndex = 10
fatIndex = 5
directory = r"C:\Users\PISUTSUNTRONKITI\Desktop\cpe341\sampling_Filtered_Nutriton.csv"
food = pandas.read_csv(directory, encoding='latin1')

def binToDec(binVar):
    return (int(binVar[0])*2**3 + int(binVar[1])*2**2 + int(binVar[2])*2**1 + int(binVar[3])*2**0)

def mutate(child):
    mutate1 = randrange(0,24) 
    mutate2 = randrange(0,24)
    while mutate2 == mutate1:
        mutate2 = randrange(0,25)
    if mutate1 != 0:
        if child[mutate1] == "0":
            child = child[0:mutate1] + '1' + child[mutate1+1:]
        else:
            child = child[0:mutate1] + '0' + child[mutate1+1:]
    else:
        if child[mutate1] == "0":
            child = '1' + child[1:]
        else:
            child = '0' + child[1:]
    if mutate2 != 0:
        if child[mutate2] == "0":
            child = child[0:mutate2] + '1' + child[mutate2+1:]
        else:
            child = child[0:mutate2] + '0' + child[mutate2+1:]
    else:
        if child[mutate2] == "0":
            child = '1' + child[1:]
        else:
            child = '0' + child[1:]
    return child        

def crossover(parent1, parent2):
    child1 = parent1[0:4] + parent2[4:8] + parent1[8:12] + parent2[12:16] + parent1[16:20] + parent2[20:24]
    child2 = parent2[0:4] + parent1[4:8] + parent2[8:12] + parent1[12:16] + parent2[16:20] + parent1[20:24]
    child1 = mutate(child1)
    child2 = mutate(child2)
    return (child1, child2)
    
def create_chromosome() :
    for i in range(32):
        gene = str(randrange(0,2))
        for j in range(23):
            gene = gene + str(randrange(0,2))
        chromosome.append(gene)
        gene = ''
    return chromosome

def findProtein(chromosome):
    meal1 = binToDec(chromosome[0:8])
    meal2 = binToDec(chromosome[8:16])
    meal3 = binToDec(chromosome[16:])
    return food.iloc[meal1-1].iloc[proteinIndex] + food.iloc[meal2-1].iloc[proteinIndex] + food.iloc[meal3-1].iloc[proteinIndex]

def findCalories(chromosome):
    meal1 = binToDec(chromosome[0:8])
    meal2 = binToDec(chromosome[8:16])
    meal3 = binToDec(chromosome[16:])
    return food.iloc[meal1-1].iloc[caloriesIndex] + food.iloc[meal2-1].iloc[caloriesIndex] + food.iloc[meal3-1].iloc[caloriesIndex]

def findSugar(chromosome):
    meal1 = binToDec(chromosome[0:8])
    meal2 = binToDec(chromosome[8:16])
    meal3 = binToDec(chromosome[16:])
    return food.iloc[meal1-1].iloc[sugarIndex] + food.iloc[meal2-1].iloc[sugarIndex] + food.iloc[meal3-1].iloc[sugarIndex]

def findCarbohydrate(chromosome):
    meal1 = binToDec(chromosome[0:8])
    meal2 = binToDec(chromosome[8:16])
    meal3 = binToDec(chromosome[16:])
    return food.iloc[meal1-1].iloc[carbIndex] + food.iloc[meal2-1].iloc[carbIndex] + food.iloc[meal3-1].iloc[carbIndex]

def findFat(chromosome):
    meal1 = binToDec(chromosome[0:8])
    meal2 = binToDec(chromosome[8:16])
    meal3 = binToDec(chromosome[16:])
    return food.iloc[meal1-1].iloc[fatIndex] + food.iloc[meal2-1].iloc[fatIndex] + food.iloc[meal3-1].iloc[fatIndex]

def checkConstraint(chromosome) :
    return findProtein(chromosome) in range(75, 100) and findSugar(chromosome) in range(24, 29) and findCalories(chromosome) in range(1500, 2000) and findCarbohydrate(chromosome) in range(150, 165) and findFat(chromosome) in range(50, 55)
        
def fitnessFn(chromosome):
    return 0.5*findProtein(chromosome) + 0.5*findCalories(chromosome)
    
def findBestAndAve(chromosome):
    for i in chromosome:
        chromosome[i] = fitnessFn(chromosome[i])
    chromosome.sort(reverse = True)
    for j in chromosome:
        crossover(chromosome[j],chromosome[j+1])

def main(): 
    chromosome = create_chromosome()
    for i in range(32):
        while !checkConstraint(chromosome[i]):
            mutate(chromosome[i])
    chromosome.sort()
    for i in range(0, 32, 2)
        childPair = crossover(chromosome[i], chromosome[i+1])
    
```
